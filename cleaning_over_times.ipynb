```python
import pandas as pd

df = pd.read_table("data/flickr_data2.csv", sep=",")

df.info()
```

```python
#columns have spaces
cols = ['id', 'user', 'lat', 'long', 'tags', 'title', 'date_taken_minute',
       'date_taken_hour', 'date_taken_day', 'date_taken_month',
       'date_taken_year', 'date_upload_minute', 'date_upload_hour',
       'date_upload_day', 'date_upload_month', 'date_upload_year',
       'Unnamed: 16', 'Unnamed: 17', 'Unnamed: 18']
df.columns = cols

df.info()
```

```python
#Some cols have dtype as object when expected numeric types
types_cols = ['date_upload_hour', 'date_upload_minute']

df[types_cols].map(type).astype(str).apply(pd.value_counts).fillna(0, downcast='infer')
```

```python
#Visualise those strings

for col in types_cols:
    print(f"Unique value in {col}: {df[col].unique()}\n")
```

```python
df.loc[df['Unnamed: 16'].isna() == False]
# df.loc[df['Unnamed: 18'].isna() == False]
```

```python
#First lets tackle the duplicates
df.duplicated()[df.duplicated() == True].sum() #232642
df = df.drop_duplicates()
df.duplicated()[df.duplicated() == True].sum() #0
df.info()
```

```python
df = df.drop(df[df['date_upload_hour'] ==  '.!'].index.item()) #One row
df = df.drop(df.loc[df['Unnamed: 18'].isna() == False].index.item()) #One row

col = "date_upload_hour"
print(f"Unique value in {col}: {df[col].unique()}\n")
df.loc[df['Unnamed: 18'].isna() == False]
```

```python
df = df.loc[df['Unnamed: 16'].isna() == True]
col = "date_upload_minute"
print(f"Unique value in {col}: {df[col].unique()}\n")
df.loc[df['Unnamed: 18'].isna() == False]
```

```python
df["date_upload_minute"] = df["date_upload_minute"].map(lambda a: int(a))
df["date_upload_hour"] = df["date_upload_hour"].map(lambda a: int(a))
print(f"Unique value in {col}: {df[col].unique()}\n")
df.loc[df['Unnamed: 18'].isna() == False]
```
